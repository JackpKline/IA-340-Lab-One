```python
df['DATE'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y') #change the data type to date with specific format
df.info()
```

```python
#rename columns
df = df.rename(columns={"NUMBER OF PERSONS KILLED": "NOP KILLED", "NUMBER OF PERSONS INJURED": "NOP INJURED","ZIP CODE":"ZC" }) 
df.head()
```

```python
df.isnull().sum()
```

#did not drop any data as it would delte my null data.

```python
df.describe() #summary of the collision data
```

```python
df = df[(df['NOP KILLED']>=0) & (df['NOP INJURED']>=0) & (df['NUMBER OF PEDESTRIANS KILLED']>=0)] #filter data
df.describe()
```

```python
df['LOCATION'].value_counts()
```

```python
df = df[['LOCATION','NOP KILLED','NOP INJURED','COLLISION_ID','NUMBER OF PEDESTRIANS KILLED','ZC']] # keep specific columns 
df.head()
```

```python
df.set_index('DATE',inplace=True) #define the index
df.head()
```

```python
df.loc['12/14/2021'] #filter the data based on the index value
```

```python
df.to_csv('clean_data.csv')
```

```python
df.hist(layout=(2,2), figsize=(10, 5), sharey=False, sharex=False, bins=50)
```

```python
sum_cases_by_date = df.groupby('DATE')['NOP KILLED'].sum() 
sum_cases_by_date.head()
```

```python
sum_cases_by_date.plot()
```

```python
sum_deaths_by_country = df.groupby('LOCATION')['NOP KILLED'].sum() # calculate the total deaths per day
sum_deaths_by_country.nlargest(10).plot.bar()
```
